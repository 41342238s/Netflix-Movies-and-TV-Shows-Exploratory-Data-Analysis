## Step 1: Load the data

````python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("netflix_titles.csv")
df.head()
````

## Step 2: Inspect and clean the data

````python
# Check data types and missing values
df.info()
print()
print("Number of missing values in each column:")
print(df.isnull().sum())
````

````python
# Check for duplicate rows
print("Number of duplicate rows:", df.duplicated().sum())
````

## Step 3: Group and analyze

````python
# Objective 1: count Movies and TV Shows for each release year
movie_counts = {}
tv_counts = {}

for i in range(len(df)):
    year = df["release_year"][i]
    show_type = df["type"][i]

    if show_type == "Movie":
        if year in movie_counts:
            movie_counts[year] = movie_counts[year] + 1
        else:
            movie_counts[year] = 1
    else:
        if year in tv_counts:
            tv_counts[year] = tv_counts[year] + 1
        else:
            tv_counts[year] = 1

# Sort by year so the line chart draws in order
movie_years = sorted(movie_counts.keys())
movie_values = []
for year in movie_years:
    movie_values.append(movie_counts[year])

tv_years = sorted(tv_counts.keys())
tv_values = []
for year in tv_years:
    tv_values.append(tv_counts[year])
````

````python
# Objective 2: split the listed_in column and count how many times each genre appears
genre_counts = {}

for i in range(len(df)):
    genres = df["listed_in"][i]
    if pd.isna(genres):
        continue
    genre_list = genres.split(", ")

    for genre in genre_list:
        if genre in genre_counts:
            genre_counts[genre] = genre_counts[genre] + 1
        else:
            genre_counts[genre] = 1

# Turn the dictionary into a list of (count, genre) pairs, then sort it
genre_count_list = []
for genre in genre_counts:
    genre_count_list.append((genre_counts[genre], genre))

genre_count_list.sort(reverse=True)
top_10_genres = genre_count_list[:10]
top_10_genres
````

## Step 4: Visualize

````python
# Line chart: number of Movies vs TV Shows by release year
plt.figure(figsize=(10, 5))
plt.plot(movie_years, movie_values, marker="o", label="Movie")
plt.plot(tv_years, tv_values, marker="o", label="TV Show")
plt.title("Number of Movies vs TV Shows by Release Year")
plt.xlabel("Release Year")
plt.ylabel("Count")
plt.legend()
plt.show()
````

````python
# Bar chart: top 10 genres
genre_names = []
genre_values = []
for count, genre in top_10_genres:
    genre_names.append(genre)
    genre_values.append(count)

# Reverse the lists so the most common genre appears at the top of the chart
genre_names.reverse()
genre_values.reverse()

plt.figure(figsize=(8, 6))
plt.barh(genre_names, genre_values, color="steelblue")
plt.title("Top 10 Genres on Netflix")
plt.xlabel("Count")
plt.ylabel("Genre")
plt.show()
````

## Step 5: Interaction

````python
## Step 5: Interaction

# Allow users to enter the year they want to search for
user_input = input("Please enter the year (before 2020) of release you wish to search for (e.g.: 2020): ")

try:
    # Convert the entered text to an integer
    search_year = int(user_input)
    
    # Filter the data for that year
    result = df[df['release_year'] == search_year]
    
    # Check for data
    if len(result) > 0:
        print(f"\n--- Search Results: {search_year} There were a total of  {len(result)} works in this year ---")
        
        # Count the number of movies and TV shows separately
        type_counts = result['type'].value_counts()
        print(f"incluging: {type_counts.get('Movie', 0)} movies, {type_counts.get('TV Show', 0)} TV Show\n")
        
        # Show the titles and descriptions of the first 10 works
        print("【10 Randomly Recommended Films from That Year】:")
        display(result[['title', 'type', 'listed_in']].head(10))
    else:
        print(f"Sorry, nothing was found in the database from {search_year} in our database.")
        
except ValueError:
    print("Invalid input format. Please make sure you entered a numeric year.（e.g.：2019）。")
````
